## 代码的解构

如何解析作业的需求？

以 0-100中的奇数为例子。
1. 打印0-100的数字
2. 判断这些数字是否为奇数
3. 输出这些数字


其中每一个都有不同的实现方式。
1. 打印0-100的数字
   1. 是直接累加
   2. 还是用list
2. 判断这些数字是否为奇数
   1. 可以取模
   2. 还是别的
3. 输出这些数字
   1. 是直接打印
   2. 还是保存到列表再打印？


下面只给出一个可能解

```python
# 这里就是使用了list
m = list(range(101))
m
```

```python
# 这是使用了循环实现了打印 和上面不同 上面是就打印了一次 一个就101行
for number in m:
    print(number)
```

```python
# 这里是对每一个数字进行判断 这里使用了取模
for number in m:
    if number % 2 == 1: # 这里的== 是判断的意思 不是=， =是赋值
        print(number) # 这里直接打印了 没有用存到某一个列表并打印（append就行）
    else:
        pass # pass就是啥也不干啦
```
